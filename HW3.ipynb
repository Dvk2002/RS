- А точно нужно сортировать по вероятности? Необязательно, если речь идет о регулярных рассылках.
Сэмплирование, а еще лучше стратифицированное сэмплирование, в рамках определенного топа позволит не надоедать клиенту одними и теми же товарами,сохраняя валидность. 

- Какую метрику использовать? Зависит от конкретных целей, но в целом, лучше взвешанную по стоимости.

- Сколько раз в неделю отпрпавляем рассылку? По моим ощущениям, чтобы ненадоедать, лучше один раз в неделю, в будни.
В выходные часто хочется отдохнуть, и любые внешние воздействия могут вызвать раздражение. Лучше отправлять в четверг-пятницу:
если у клиента действительно есть интерес, он может вернуться к нему на выходных, когда будет время, с другой стороны,
оно не затеряется среди остальных сообщений, отправленных в течение недели.

- В какое время отправляем рассылку? Не нужно отправлять в экстремальные периоды, лучше не очень ранним утром, или после
обеда

- Будем отправлять одному юзеру много раз наши рекоммендации. Как добиться того, чтобы они хоть немного отличались? см. п1

- Нужно ли, чтобы в одной рассылке были *разные* товары? Как определить, что товары *разные*? Как добиться того, чтобы они были разными? Нужно классифицировать и фильтровать по категориям (ввести ограничение на количество товаров определенной категории). Более того, однообразие мало кому нравится, и есть смысл также варьировать методы рассылки одному и тому же клиенту: одну неделю смс, вторую - email, третью push и т.д.
